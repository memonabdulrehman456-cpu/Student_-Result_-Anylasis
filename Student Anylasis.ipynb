```python
import pandas as pd

# =========================================
# Student Result Analysis Project
# Author: Abdul Rehman Memon
# =========================================

# Step 1: Load student data
students_data = pd.read_csv("students.csv")

# Step 2: Define subjects list
subjects = ["Math","Science","English","Urdu","Islamiyat","PST","Computer","Arts"]

# Step 3: Calculate total marks
students_data["Total_Marks"] = students_data[subjects].sum(axis=1)

# Step 4: Calculate percentage
students_data["Percentage"] = students_data["Total_Marks"] / len(subjects)

# Step 5: Function to assign grade
def calculate_grade(percent):
    if percent >= 90:
        return "A+"
    elif percent >= 80:
        return "A"
    elif percent >= 70:
        return "B"
    elif percent >= 60:
        return "C"
    else:
        return "Fail"

# Step 6: Apply grade to each student
students_data["Grade"] = students_data["Percentage"].apply(calculate_grade)

# Step 7: Show Top 10 students
top_students = students_data.sort_values("Percentage",ascending=False).head(20)
print(" Top 10 Students:")
print(top_students)

# Step 8: Save final results
students_data.to_csv("final_results.csv", index=False)
print(" Final results saved as final_results.csv")
```
